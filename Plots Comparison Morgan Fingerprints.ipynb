```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("morgan_vs_polyAGM.csv")
df
```

```python
# Data
df1 = df[0:5]  # Binary Morgan Fingerprints
df2 = df[5:10]  # Count Morgan Fingerprints
df3 = df[10:15]  # PolyAGM (WL) - Weisfeiler-Lehman 
df4 = df[15:20]  # PolyAGM (WL + SP) - Weisfeiler-Lehman and Shortest Paths
```

```python
def plot_property(prop):

    # multiple line plots
    plt.plot( range(1,6), prop, data=df1, marker='o', color='red', linewidth=2, linestyle='dashed', label= "Morgan Fingerprints (Binary)")
    plt.plot( range(1,6), prop, data=df2, marker='o', color='orange', linewidth=2, linestyle='dashed', label="Morgan Fingerprints (Count)")
    plt.plot( range(1,6), prop, data=df3, marker='o', color='blue', linewidth=2, linestyle='dashed', label="PolyAGM (WL)")
    plt.plot( range(1,6), prop, data=df4, marker='o', color='purple', linewidth=2, linestyle='dashed', label="PolyAGM (WL + SP)")

    plt.xticks(range(1,6))
    plt.xlabel("Depth of Motifs")
    plt.ylabel("Error in predicting " + pretty_target_names[prop])
    # show legend
    plt.title("Error when predicting " + pretty_target_names[prop] + " (lower is better)")
    plt.legend()

    # show graph
#     plt.show()
    
    # save graph
    plt.savefig("Error when predicting " + prop + ".svg", format="svg", dpi=2400)
    plt.savefig("Error when predicting " + prop + ".pdf", format="pdf", dpi=2400)
```

```python
pretty_target_names = {
        'Tg' : 'Lower $T_g$',
         'Tg2': 'Upper $T_g$',
         'σbreak (MPa)' : '$σ_{break}$ (MPa)',
         'εbreak (pct)' : '$ε_{break}$ (%)',
}
```

```python
plot_property('εbreak (pct)')
```

```python
plot_property('σbreak (MPa)')
```

```python
plot_property('Tg')
```

```python
plot_property('Tg2')
```
